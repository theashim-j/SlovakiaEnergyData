```python
import pandapower.plotting as pplot
import matplotlib.pyplot as plt

# Define categories of generator and slack buses
bus_color_map = {
    'Slack': ['Malzenice Gas'],
    'Nuclear': ['Bohunice Nuclear', 'Mochovce Nuclear', 'Nove Zamky'],
    'Hydro': ['Cierny Vah', 'Gabcikova', 'Liptovska Mara Gen'],
    'Solar': ['Komarno', 'Nitra'],
    'Wind': ['Skalite', 'Cerova', 'Myjava'],
    'Biomass': ['Bratislava', 'Kosice', 'Presov'],
    'Interconnector':['Czech','Poland','Ukraine']
}

# Define default color for load buses
default_load_color = 'gray'

# Assign colors to each generator type
bus_type_colors = {
    'Slack': 'red',
    'Nuclear': 'orange',
    'Hydro': 'aqua',
    'Solar': 'gold',
    'Wind': 'green',
    'Biomass': 'brown',
    'Interconnector' : 'blue',
    'Load': default_load_color
}


# Create a mapping of bus names to categories
bus_to_type = {}
for category, buses in bus_color_map.items():
    for bus in buses:
        bus_to_type[bus] = category

# Extract bus coordinates and names
bus_coords = Slovakia_Grid_Net.bus_geodata[['x', 'y']].values
bus_names = Slovakia_Grid_Net.bus['name'].values

# Plot setup
plt.figure(figsize=(14, 10))
ax = plt.gca()

# Plot buses with customized colors
for coord, name in zip(bus_coords, bus_names):
    bus_type = bus_to_type.get(name, 'Load')  # Default to 'Load' if not categorized
    bus_color = bus_type_colors[bus_type]

    # Plot bus points
    plt.plot(coord[0], coord[1], 'o', color=bus_color, markersize=8)

    # Annotate bus names
    plt.text(coord[0], coord[1], name, fontsize=8, ha='center', va='bottom', rotation=20)

# Remove axis labels, grids, and frames
plt.axis('off')

# Adjust margins manually
plt.subplots_adjust(left=0.01, right=0.99, top=0.99, bottom=0.01)

# Optional: Add legend
for bus_type, color in bus_type_colors.items():
    plt.plot([], [], 'o', color=color, label=bus_type)
plt.legend(loc='best', fontsize=9, frameon=False)

plt.show()
```

```python
import pandapower.plotting as pplot
import matplotlib.pyplot as plt

# Define categories of generator and slack buses
bus_color_map = {
    'Slack': ['Malzenice Gas'],
    'Nuclear': ['Bohunice Nuclear', 'Mochovce Nuclear', 'Nove Zamky'],
    'Hydro': ['Cierny Vah', 'Gabcikova', 'Liptovska Mara Gen'],
    'Solar': ['Komarno', 'Nitra'],
    'Wind': ['Skalite', 'Cerova', 'Myjava'],
    'Biomass': ['Bratislava', 'Kosice', 'Presov'],
    'Interconnector':['Czech','Poland','Ukraine']
}

# Define default color for load buses
default_load_color = 'gray'

# Assign colors to each generator type
bus_type_colors = {
    'Slack': 'red',
    'Nuclear': 'orange',
    'Hydro': 'aqua',
    'Solar': 'gold',
    'Wind': 'green',
    'Biomass': 'brown',
    'Interconnector' : 'blue',
    'Load': default_load_color
}

# Create reverse lookup: bus name → category
bus_to_type = {}
for category, names in bus_color_map.items():
    for name in names:
        bus_to_type[name] = category

# Extract geodata for plotting
bus_coords = Slovakia_Grid_Net.bus_geodata[['x', 'y']].values
bus_names = Slovakia_Grid_Net.bus['name'].values

# Begin plotting
plt.figure(figsize=(14, 10))
ax = plt.gca()

for coord, name in zip(bus_coords, bus_names):
    bus_type = bus_to_type.get(name, 'Load')  # Default to Load if not found
    color = bus_type_colors[bus_type]

    plt.plot(coord[0], coord[1], 'o', color=color, markersize=8)
    plt.text(coord[0], coord[1], name, fontsize=10, ha='center', va='bottom', rotation=20)

# Clean up plot aesthetics
plt.axis('off')
plt.subplots_adjust(left=0.01, right=0.99, top=0.99, bottom=0.01)


# Create and display legend
for category, color in bus_type_colors.items():
    plt.plot([], [], 'o', color=color, label=category)
plt.legend(loc='best', fontsize=9, frameon=False)

plt.title("Slovakia Grid: Generator and Load Bus Map", fontsize=14)

# Draw lines between buses
line_df = Slovakia_Grid_Net.line
for _, line in line_df.iterrows():
    fb = line['from_bus']
    tb = line['to_bus']
    x_coords = [Slovakia_Grid_Net.bus_geodata.loc[fb, 'x'], Slovakia_Grid_Net.bus_geodata.loc[tb, 'x']]
    y_coords = [Slovakia_Grid_Net.bus_geodata.loc[fb, 'y'], Slovakia_Grid_Net.bus_geodata.loc[tb, 'y']]
    
    plt.plot(x_coords, y_coords, color='black', linewidth=1, alpha=0.6)  # customize style here

plt.show()
```

```python
import matplotlib.pyplot as plt

# Line results
line_results = Slovakia_Grid_Net.res_line

# Real power losses in MW
line_results['pl_mw']  # This shows the losses on each line

# Reactive power losses in MVAr
line_results['ql_mvar']  # This shows the losses on each line

# Sum of total losses in the network
total_real_loss = line_results['pl_mw'].sum()
total_reactive_loss = line_results['ql_mvar'].sum()

print("Total Real Power Loss (MW):", total_real_loss)
print("Total Reactive Power Loss (MVAr):", total_reactive_loss)

# Extract necessary values
p_from = line_results['p_from_mw']  # Power sent from "from" bus
p_loss = line_results['pl_mw']      # Real power loss

# Calculate percentage loss
line_results['percentage_loss'] = abs((p_loss / p_from) * 100)

# Print the results clearly
print(line_results[['p_from_mw', 'pl_mw', 'percentage_loss']])

# Plot explicitly using matplotlib figure and axis objects
fig, ax = plt.subplots(figsize=(12, 7))
line_results['percentage_loss'].plot(kind='bar', ax=ax)

# Customize plot clearly
ax.set_title('Power Loss in Each Line')
ax.set_xlabel('Line Index')
ax.set_ylabel('Percentage Loss (%)')
plt.grid(axis='y')

# Save the figure explicitly
fig.savefig("Power_loss.png", dpi=1000)

# Show the plot
plt.show()
```
